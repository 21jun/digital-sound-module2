```python
from pathlib import Path
import numpy as np
import scipy.io.wavfile as wf
from scipy import spatial
import matplotlib.pyplot as plt
```

```python
train_path = Path("./voice")
get= train_path.glob("**/*.wav")
files = [x for x in get if x.is_file()]

# test_path = Path("./target")
test_path = Path("./test")
test_files = [x for x in test_path.glob("**/*.wav") if x.is_file()]
```

```python
def get_fft_data(data, st):
    d = data[st*16000 : st*16000 + 4*16000]
    ft = np.fft.fft(d)
    return np.sqrt(ft.real**2 + ft.imag**2)[:int(d.shape[0]/2)]


def get_4sec_frame(data, st, normalize= False):
    d = data[int(st*16000) : int(st*16000) + 4*16000]
    if normalize:
        d = d/2.0
    return d


def get_4sec_fft_data(data):
    d = data
    ft = np.fft.fft(d)
    return np.sqrt(ft.real**2 + ft.imag**2)[:int(d.shape[0]/2)]

def predict(label, fft_voice, mean_fft_voices):
    
    prediction = {}
    for name, data in mean_fft_voices.items():
        cos = 1 - spatial.distance.cosine(fft_voice, data)
        prediction[name] = cos

    prediction_desc = sorted(prediction.items(), key=lambda kv: kv[1], reverse=True)
    # print(f"{label} 유사도 ... gap :", prediction_desc[0][1] - prediction_desc[1][1])
    print(f"{label} 의 최종 예측 Label =", prediction_desc[0][0], end='\n\n')
    # if label.split('_')[0] != prediction_desc[0][0].split('.')[0]:
    #     print("wrong!")
    
def predict_with_starting_time(name, st, normalize=False):
    wave = get_4sec_frame(voices[name], st, normalize)
    fft  = get_4sec_fft_data(wave)
    predict(name, fft, mean_fft_voices)
    
    
def plot_wave_fft(name='정유라.wav', normalize=False ):
    fig, axs = plt.subplots(2, figsize=(10,20))

    for i in range(0,8,2):

        wave = get_4sec_frame(voices[name], i, normalize)
        fft  = get_4sec_fft_data(wave)
        axs[0].plot(wave, alpha=0.5)
        axs[1].plot(fft, alpha = 0.5)
```

```python
voices = {}
for i, file in enumerate(files):
    try:
        _, data = wf.read(file)
        voices[file.name] = data
    except:
        print(f"can't read {file.name}")
```

```python
fft_voices = {}
for name, data in voices.items():
    fft_voices[name] = [get_fft_data(data, x) for x in range(10)]

mean_fft_voices = {}
for name, data in fft_voices.items():
    mean_fft_voices[name] = np.array(data).mean(axis=0)
```
